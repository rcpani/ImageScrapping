```python
#store all in a variable called image_tags
image_tags=soup.find_all("img")
del image_tags[0]
image_tags
```

```python
save_dir="images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
for i in image_tags:
    print(image_tags.index(i))
```

```python
# to fetch all the links
for i in image_tags:
    #get urls
    image_url=(i['src'])
    #request to get the urls
    image_data=requests.get(image_url).content
    #with open(os.path.join(save_dir,f"{query}_{image_tags.index(i)}"))
    with open(os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpg"), "wb") as f:
              f.write(image_data)
```
